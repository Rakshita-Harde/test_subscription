**1. What is the distribution of users across different subscription plans?**

```python
## User distribution across different subscription plans 
plt.figure(figsize=(6,4))
sns.countplot(x='Subscription_Plan', data=df, palette='viridis')
plt.title('User Distribution Across Subscription Plans')
plt.xlabel('Subscription Plan')
plt.ylabel('Number of Users')
plt.show()
```

- **As seen in the plot Standard plan is most famous plan amongs the users**
- **Then the basic plan is almost same as the standard plan**
- **Preminun plan is least famous as it is costly then other plans**


**2. How do engagement scores vary between users of different plans?**

```python
## Engagement Score Vs subscribstion plan 

plt.figure(figsize=(6,4))
sns.boxplot(x='Subscription_Plan', y='Engagement_Score', data=df, palette='Greens')
plt.title('Engagement Score by Subscription Plan')
plt.show()
```

- **Coustomers with basic plans has less engagement score among all as this plan has less Amount of Gb that could be a reason**
- **Coustomers with standard plans has average engagement score**
- **Coustomers with premium plan has highest engagement score among all as this types of data plans has more amount of 
    data the usage also increases**

```python
cat_col = ['Payment_Mode','City_Tier']
plt.figure(figsize=(12, 13))
for i, col in enumerate(cat_col, 1):
    plt.subplot(3, 2, i)
    sns.countplot(data=df, x=col, palette='Set1')
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

1. **These graphs shoews the count of payment methods used by coustomers ,mostly used method is creditcard,then UPIs ,few uses wallets and remaing
 uses autodebit**
2. **Metro cities has highest count in subscriptions , also the tier1 city has good count of subscriptions
   comparitively less in tier2 cities  and tier3 cities has the least counts**
